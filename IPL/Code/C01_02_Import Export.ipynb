```python

for filenum in np.arange(total_files):

    import_filename = files_only[filenum].replace('.json','')
    print(import_filename)
    
    if (import_filename != 'README.txt') :    
        if int(import_filename) >= 1359475:
            # Replace 'file.json' with the path to your JSON file
            json_file_path = location + '\\' + files_only[filenum]
            print(json_file_path)



            #adding a check to see if json file is empty before reading it with pd.read_json.

            if os.path.getsize(json_file_path) > 0:
                # Can't read it directly because of multiple structures inside the json file
                #df = pd.read_json(json_file_path)

                # let's try to flatten it and then import it
                with open(json_file_path, 'r', encoding='utf-8') as file:
                    json_content = json.load(file)

                # Flatten the nested JSON content
                df = pd.json_normalize(json_content)

            else:
                print("The JSON file is empty.")

            i_ss = json_content['info']['season']
            i_mtype = json_content['info']['match_type']
            i_date = json_content['info']['dates'][0].replace('-','_')
            i_city = json_content['info']['city']
            i_tms = json_content['info']['teams'][0] + '_v_' + json_content['info']['teams'][1]
            
            i_match_num = json_content['info']['event']['match_number']

            info_ = [i_match_num, i_ss, i_mtype, i_date, i_city, i_tms ]

            print(info_)
            #json_content['info']['dates'].pop()

            tms_lst = json_content['info']['teams']

            print(tms_lst)




            # looping through the json file(s) to put into dataframe

            colslist = ['mnum','inn','bat_tm','bat','ball','bow','runs','wkt','wkt_how']

            df_innings = pd.DataFrame(columns = colslist)


            for innings in [0,1]:
                print(type(json_content['innings'][innings]['overs']))
                overs_in_this_inn = len(json_content['innings'][innings]['overs'])

                bat_tm = json_content['innings'][innings]['team']
                print(f'Over in Innings is {overs_in_this_inn}')
                for over_num in np.arange(0,overs_in_this_inn,1):
                    #print(len(json_content['innings'][innings]['overs'][over_num]['deliveries']))

                    balls_in_over = len(json_content['innings'][innings]['overs'][over_num]['deliveries'])
                    for ball_num in np.arange(0,balls_in_over,1):
                        wkt = 0
                        wkt_how = ''
                        ball = 1

                        #print(len(json_content['innings'][innings]['overs'][over_num]['deliveries'][ball_num]))
                        bat_on_str = json_content['innings'][innings]['overs'][over_num]['deliveries'][ball_num]['batter']
                        bow = json_content['innings'][innings]['overs'][over_num]['deliveries'][ball_num]['bowler']
                        runs_to_bat = json_content['innings'][innings]['overs'][over_num]['deliveries'][ball_num]['runs']['batter']

                        if 'wickets' in json_content['innings'][innings]['overs'][over_num]['deliveries'][ball_num]:
                            wkt = 1
                            wkt_how = json_content['innings'][innings]['overs'][over_num]['deliveries'][ball_num]['wickets'][0]['kind']
                            
                        if 'extras' in json_content['innings'][innings]['overs'][over_num]['deliveries'][ball_num]:
                            if ('noballs' in json_content['innings'][innings]['overs'][over_num]['deliveries'][ball_num]['extras']) or ('wides' in json_content['innings'][innings]['overs'][over_num]['deliveries'][ball_num]['extras']):
                                ball = 0

                        this_ball_details_row = [i_match_num, innings+1, bat_tm, bat_on_str, ball, bow, runs_to_bat, wkt, wkt_how]
                        #print(this_ball_details_row)
                        df_innings.loc[len(df_innings)] = this_ball_details_row


            df_innings['four'] = df_innings['runs'].apply(lambda x : 1 if x==4 else 0)
            df_innings['six'] = df_innings['runs'].apply(lambda x : 1 if x==6 else 0)

            df_innings['bow_tm'] = df_innings.apply(lambda x: next(item for item in tms_lst if item != x.bat_tm), axis=1)

            df_innings.head()




            #batter rollup

            df_bat_stats = df_innings.groupby(['bat_tm','bow_tm','inn','bat','mnum'])\
            .agg({'runs':'sum','four':'sum','six':'sum','wkt':'sum', 'ball':'sum'})\
                .reset_index().rename(columns = {'runs':'rs','ball':'bowf','wkt':'out','inn':'inn_bat','bat':'name'})

            print(df_bat_stats.columns.values.tolist())
            print(df_bat_stats.shape)

            #df_bat_stats['bow_tm'] = df_bat_stats.apply(lambda x : [item for item in two_member_list if item != character_variable] , axis = 1)
            #df_bat_stats['bow_tm'] = df_bat_stats.apply(lambda x : item for item in tms_list if item != x.bat_tm , axis = 1)

            display(df_bat_stats.head())

            #batter rollup

            df_bow_stats = df_innings.groupby(['bow','bow_tm','bat_tm','inn','mnum']).agg({'runs':'sum', 'ball':'sum','wkt':'sum'})\
                .reset_index().rename(columns = {'runs':'rc','ball':'balb','inn':'inn_bow','bow':'name'})

            print(df_bow_stats.columns.values.tolist())
            print(df_bow_stats.shape)

            df_bow_stats['inn_bat'] = df_bow_stats['inn_bow'].apply(lambda x : 2 if x==1 else 1 )

            display(df_bow_stats.head())




            print(df_bow_stats.shape)
            print(df_bat_stats.shape)

            df_match = pd.merge( df_bat_stats, df_bow_stats\
                                , left_on = ['bow_tm','bat_tm','inn_bat','name','mnum']
                                , right_on = ['bat_tm','bow_tm','inn_bat','name','mnum']
                                , how = 'outer'
                                , suffixes=('_inn1', '_inn2') ) 

            print(df_match.shape)


            display(df_match.sort_values(by = ['inn_bat']))


            df_match['season'] = i_ss
            df_match['type'] = i_mtype
            df_match['date'] = i_date
            df_match['city'] = i_city
            df_match['teams'] = i_tms
            
            df_match['bat_tm_inn1'] = df_match['bat_tm_inn1'].astype(str)
            df_match['bow_tm_inn2'] = df_match['bow_tm_inn2'].astype(str)
            
            df_match['team_nm'] = df_match.apply(lambda x : x.bow_tm_inn2 if ( x.bat_tm_inn1 == 'nan') else x.bat_tm_inn1, axis = 1)
            #df_match['team_nm'] = df_match['bat_tm_inn1'] + df_match['bow_tm_inn2']

            display(df_match.head())

            export_fname = import_filename + '_' + i_mtype + '_' + str(i_ss) + '_' + str(i_match_num) + '_' + i_date + '_' + i_tms + '.csv'

            print(export_fname)
            
            dropcols = ['bat_tm_inn1','bat_tm_inn2','bow_tm_inn1','bow_tm_inn2']
            df_match_exp = df_match.sort_values(by='inn_bat').drop(columns = dropcols)

            df_match_exp.to_csv( exp_location + '\\' + export_fname, index = False )


```
