```python

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python

loc_exp = "C:\\Users\\karan\\Documents\\Cricket-Analytics\\IPL\\Data\\exp\\"

fname_gstats = 'gstats_fyaml_G70.xlsx'
```

```python

dfs = pd.read_excel( loc_exp + fname_gstats, sheet_name=None)

# dfs will be a dictionary with sheet names as keys and dataframes as values.
# You can access a specific sheet's dataframe like this:

for keys in dfs.keys():
    print(keys)
```

```python

df_comb = dfs['Combined']

print(df_comb.shape)
print()
print(df_comb.columns.values.tolist())
print()

display(df_comb.head(25))
```

```python

# Assuming your data is in a pandas DataFrame df
# Select only numeric columns for clustering
numeric_columns = ['tbinn', 'tr', 'tr_pp', 'tr_mid', 'tr_dth', 'tbov', 'tbov.spin', 'tbov.pace', 'tout', 'tout.spin', 'tout.pace', 'trc']

#df_numeric = df_comb[numeric_columns]

# assuming df is your original dataframe
new_df = pd.DataFrame()

for column in numeric_columns:
    if column != 'tbinn':
        new_df[column + "_tbinn_ratio"] = df_comb[column] / df_comb['tbinn']


new_df['city'] = df_comb['city'] 
```

```python

print(new_df.columns.values.tolist())
print()
print(new_df.shape)
print()
display(new_df.head(25))
```

```python

numeric_columns = ['tr_tbinn_ratio', 'tr_pp_tbinn_ratio', 'tr_mid_tbinn_ratio', 'tr_dth_tbinn_ratio', 'tbov_tbinn_ratio', 'tbov.spin_tbinn_ratio', 'tbov.pace_tbinn_ratio', 'tout_tbinn_ratio', 'tout.spin_tbinn_ratio', 'tout.pace_tbinn_ratio', 'trc_tbinn_ratio']

new_df_numeric = new_df[numeric_columns]

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(new_df_numeric)

# Fit KMeans
kmeans = KMeans(n_clusters=7, random_state=0).fit(data_scaled)  # You might need to adjust n_clusters based on your data

# Add the cluster labels to the original DataFrame
new_df['cluster'] = kmeans.labels_

# Group by city and find the most common cluster for each city
city_clusters = new_df.groupby('city')['cluster'].agg(lambda x: x.value_counts().index[0])

print(city_clusters)

```

```python

wcss = kmeans.inertia_
print('Within-cluster sum of squares (WCSS):', wcss)
```

```python

from sklearn.metrics import silhouette_score

score = silhouette_score(data_scaled, kmeans.labels_)
print('Silhouette Score:', score)
```
