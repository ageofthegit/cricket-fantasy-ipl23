```python

groupbycols = ['tm_bat', 'bat', 'm.num','bat_order']

#bat_grouped_df = master_df.groupby(['bat_tm', 'batnum', 'bat', 'm.num'])
bat_grouped_df = master_df_tags[ (master_df_tags['b.num'] != 0.0) & ~(master_df_tags['inning'].isnull()) ].groupby(groupbycols) # batnum unique as the m.num per player

def custom_agg(group):
    data = {
        'tr': group['tr'].sum(),
        'tout': group['wkt'].sum(),
        'tinn': group['m.num'].nunique(),
        'tbf': group[group['b.num'] != 0.0]['b.num'].count(),
                
        'tr_pp': (group['fpp'] * group['tr']).sum(),
        'tr_mid': (group['fmid'] * group['tr']).sum(),
        'tr_dth': (group['fdth'] * group['tr']).sum(),
        
        'tout_pp': (group['wkt'] * group['fpp']).sum(),        
        'tout_mid': (group['wkt'] * group['fmid']).sum(),        
        'tout_dth': (group['wkt'] * group['fdth']).sum(),
        
        'tbf_pp': group[ (group['b.num'] != 0.0) ]['fpp'].sum(),
        'tbf_mid': group[ (group['b.num'] != 0.0) ]['fmid'].sum(),
        'tbf_dth': group[ (group['b.num'] != 0.0) ]['fdth'].sum(),
        
        'tr_pace': group[group['bowlcat'] == 'Pace']['tr'].sum(),
        'tr_spin': group[group['bowlcat'] == 'Spin']['tr'].sum(),
        
        'tout_pace': group[group['bowlcat'] == 'Pace']['wkt'].sum(),
        'tout_spin': group[group['bowlcat'] == 'Spin']['wkt'].sum()
        
    }
    return pd.Series(data)

batting_match_level_df = bat_grouped_df.apply(custom_agg)

# Reset the index so that the primary key columns become regular columns
batting_match_level_df.reset_index(inplace=True)
```

```python

print(batting_match_level_df.shape)

df_bat_transpose = batting_match_level_df.copy()
df_bat_transpose['game.num'] = 'G' + df_bat_transpose['m.num'].astype(str).str.zfill(2)

print(df_bat_transpose.shape)

df_bat_transpose.sort_values(by = 'm.num', ascending = False)

batting_gwise = df_bat_transpose.pivot(index= ['tm_bat','bat'], columns= ['game.num'], values= ['tr']).reset_index()

# Fixing multi-level columns
#batting_gwise.columns = batting_gwise.columns.map(lambda x: f'{x[1]}.{x[0]}')

# Flatten only multi-level column index
batting_gwise.columns = batting_gwise.columns.map( lambda x: f'{x[0]}.{x[1]}' if len(x) > 1 else x[0] )

# Reset index
batting_gwise = batting_gwise.reset_index(drop = True).rename(columns = {'tm_bat.':'tm_bat','bat.':'bat'})
```

```python

# Group by 'tm_bat', 'bat', 'bat_order' and count the occurrences
bat_order_counts = df_bat_transpose.groupby(['tm_bat', 'bat', 'bat_order']).size().reset_index(name='counts')


# Find the batting order with maximum count for each batsman
most_batted_position_idx = bat_order_counts.groupby(['tm_bat', 'bat'])['counts'].idxmax()

# Create a new DataFrame for most_batted_position using the idxmax result
most_batted_position = bat_order_counts.loc[most_batted_position_idx, ['tm_bat', 'bat', 'bat_order']].rename(columns={'bat_order': 'most_batted_position'})

# Merge most_batted_position into bat_order_counts
bat_order_counts = bat_order_counts.merge(most_batted_position, on=['tm_bat', 'bat'], how='left')


# Create a new column 'bat_order_counts' in the format 'bat_order(counts)'
bat_order_counts['bat_order_counts'] = bat_order_counts['bat_order'].astype(str) + "(" + bat_order_counts['counts'].astype(str) + ")"

# Group by 'tm_bat' and 'bat' and join all 'bat_order_counts' together separated by ', '
bat_order_counts = bat_order_counts.groupby(['tm_bat', 'bat']).agg({'bat_order_counts': ', '.join, 'most_batted_position': 'first'}).reset_index()

# Now merge this DataFrame with your original one
batting_gwise = pd.merge(batting_gwise, bat_order_counts, on=['tm_bat', 'bat'], how='left')

```
