```python

# Get the HTML content of the website
response = requests.get('https://www.espncricinfo.com/series/indian-premier-league-2023-1345038/chennai-super-kings-vs-kolkata-knight-riders-61st-match-1359535/ball-by-ball-commentary')

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, 'html.parser')
soup_innings = BeautifulSoup(response.content, 'html5lib')

# pprint soup to see what is going on here
pprint(soup_innings)
```

```python

# Find all the ball-by-ball commentary elements
commentary_elements = soup.find_all('div', class_='commentary')

pprint(commentary_elements)
```

```python
# Extract the data from the commentary elements
data = []
for commentary_element in commentary_elements:
    # Extract the over number
    over_number = commentary_element.find('span', class_='over').text

    # Extract the ball number
    ball_number = commentary_element.find('span', class_='ball').text

    # Extract the batsman
    batsman = commentary_element.find('span', class_='batsman').text

    # Extract the bowler
    bowler = commentary_element.find('span', class_='bowler').text

    # Extract the delivery type
    delivery_type = commentary_element.find('span', class_='delivery-type').text

    # Extract the runs scored
    runs_scored = commentary_element.find('span', class_='runs').text

    # Extract the wickets taken
    wickets_taken = commentary_element.find('span', class_='wickets').text

    # Extract the extras
    extras = commentary_element.find('span', class_='extras').text

    # Extract the total runs
    total_runs = commentary_element.find('span', class_='total-runs').text

    # Append the data to the list
    data.append({
        'over_number': over_number,
        'ball_number': ball_number,
        'batsman': batsman,
        'bowler': bowler,
        'delivery_type': delivery_type,
        'runs_scored': runs_scored,
        'wickets_taken': wickets_taken,
        'extras': extras,
        'total_runs': total_runs
    })

```

```python


# Print the data
print(data)

```

```python

import requests
import json
import pandas as pd
from pprint import pprint
```

```python

def scrape_chunk(series_id, match_id, from_over, inning_num):
    """
    scrapes a chunk of commentary via AJAX request. The commentary comes in two over pieces, going backwards
    from the from_over parameter 
    e.g if from_over = 2 then we will get overs 1 and 2 which will have ball details looking like 0.1, 0.2, 1.1, 1.2 etc
    """
    url = f"https://hs-consumer-api.espncricinfo.com/v1/pages/match/comments?lang=en&seriesId={series_id}&matchId={match_id}&inningNumber={inning_num}&commentType=ALL&fromInningOver={from_over}"
    #print(url)
    resp = requests.get(url)
    resp_json = json.loads(resp.text)
    chunk_data = resp_json['comments']
    return chunk_data

```

```python

G61_url = "https://www.espncricinfo.com/series/indian-premier-league-2023-1345038/chennai-super-kings-vs-kolkata-knight-riders-61st-match-1359535/ball-by-ball-commentary"

#series_id = 1239527 
series_id = 1345038

#match_id = 1239543
match_id = 1359535

end_over = 20
inning_num = 1

two_overs = scrape_chunk(series_id, match_id, end_over, inning_num)

print(type(two_overs))
print(len(two_overs))

# len means that there are x number of balls in that two-over chunk ending at from_over
```

```python

#bbb_colname = ['inn','onum','bnum','tr', 'trb', 'ffour', 'fsix', 'fwkt', 'out.type', 'pid.bat', 'pid.bow', 'inn.tr', 'inn.wkt', 'tm.bat']
bbb_colname = ['inn','onum','bnum','tr', 'trb', 'ffour', 'fsix', 'fwkt', 'out.type', 'pid.bat', 'pid.bow', 'inn.tr', 'inn.wkt']

bbb_master_df = pd.DataFrame(columns = bbb_colname)


for inning_num in np.arange(1,3):
    for end_over in np.arange(2,22,2):    
        print(f'Getting data for inning_num : {inning_num}, end_over : {end_over}')

        two_overs = scrape_chunk(series_id, match_id, end_over, inning_num)


        for ball_index in np.arange(len(two_overs)):

            #print(two_overs[ball_index]['inningNumber'])
            inn = two_overs[ball_index]['inningNumber']

            #print(two_overs[ball_index]['overNumber'])
            onum = two_overs[ball_index]['overNumber']

            #print(two_overs[ball_index]['ballNumber'])
            bnum = two_overs[ball_index]['ballNumber']

            #print(two_overs[ball_index]['totalRuns'])
            tr = two_overs[ball_index]['totalRuns']
            trb = two_overs[ball_index]['batsmanRuns']
            #print(two_overs[ball_index]['id'])

            ffour = two_overs[ball_index]['isFour']
            fsix = two_overs[ball_index]['isSix']
            fwkt = two_overs[ball_index]['isWicket']
            out_type = two_overs[ball_index]['dismissalType']

            pid_bat = two_overs[ball_index]['batsmanPlayerId']
            pid_bow = two_overs[ball_index]['bowlerPlayerId']

            inn_tr = two_overs[ball_index]['totalInningRuns']
            inn_wkt = two_overs[ball_index]['totalInningWickets']

            #tm_bat = two_overs[ball_index]['over']['team']['abbreviation']

            bbb_rowlist = [inn, onum, bnum, tr, trb, ffour, fsix, fwkt, out_type, pid_bat, pid_bow, inn_tr, inn_wkt]
            bbb_master_df.loc[len(bbb_master_df)] = bbb_rowlist
            
print(bbb_master_df.shape)

```
