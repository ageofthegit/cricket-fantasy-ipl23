```python

# Defining the transition matrix using the 'state' variable

def create_transition_matrix(data, states, state_to_index):
    transition_matrix = np.zeros((len(states), len(states)))

    for i in range(len(data) - 1):
        current_state = data.iloc[i]['state']
        next_state = data.iloc[i+1]['state']
        transition_matrix[state_to_index[current_state], state_to_index[next_state]] += 1

    row_sums = transition_matrix.sum(axis=1, keepdims=True)
    transition_matrix = transition_matrix / np.where(row_sums != 0, row_sums, 1)

    return transition_matrix
```

```python
def simulate_ball_by_ball(transition_matrix, state_to_index, states, batting_order):
    current_state = 0
    simulation = []
    runs = 0
    wickets = 0
    batsman_index = 0

    for ball in range(len(batting_order) * 6):  # Each batsman faces 6 balls
        # End the simulation if all wickets have fallen
        if batsman_index >= len(batting_order):
            break

        next_state_probs = transition_matrix[state_to_index[current_state]]
        next_state_index = np.random.choice(range(len(states)), p=next_state_probs)
        next_state = states[next_state_index]
        simulation.append((batting_order[batsman_index], next_state))
        current_state = next_state
        runs += next_state

        if next_state == 99:  # Wicket
            wickets += 1
            batsman_index += 1

        if ball % 6 == 0 and ball != 0:  # End of over
            batsman_index = (batsman_index + 1) % len(batting_order)

    return simulation, runs, wickets
```

```python

def simulate_ball_by_ball(transition_matrix, state_to_index, states, batting_order, target ):
    current_state = 0
    simulation = []
    runs = 0
    wickets = 0
    batsman_index = 0
    ball_number = 0

    while batsman_index < len(batting_order):
        next_state_probs = transition_matrix[state_to_index[current_state]]
        next_state_index = np.random.choice(range(len(states)), p=next_state_probs)
        next_state = states[next_state_index]
        simulation.append((batting_order[batsman_index], next_state))
        current_state = next_state
        runs += next_state if next_state != 99 else 0
        ball_number += 1

        if next_state == 99 or ball_number == 120:  # A wicket fell or over completed
            batsman_index += 1
            ball_number = 0  # reset ball count for the next batsman

        if runs >= target or wickets == 10:  # Stop if target is reached or all out
            break

    return simulation, runs, wickets
```

```python

def create_scorecard(simulation):
    scorecard = defaultdict(lambda: {'runs': 0, 'balls': 0, 'out': False})

    for batsman, runs in simulation:
        scorecard[batsman]['runs'] += runs if runs != 99 else 0
        scorecard[batsman]['balls'] += 1
        if runs == 99:
            scorecard[batsman]['out'] = True

    # Convert to DataFrame
    scorecard_df = pd.DataFrame(scorecard).T
    scorecard_df.index.name = 'Batsman'
    scorecard_df.reset_index(inplace=True)

    return scorecard_df

# Create scorecards for both innings
scorecard_first_innings = create_scorecard(simulated_first_innings)
scorecard_second_innings = create_scorecard(simulated_second_innings)

# Print the scorecards
print("First Innings Scorecard:")
print(scorecard_first_innings)
print("\nSecond Innings Scorecard:")
print(scorecard_second_innings)
```
