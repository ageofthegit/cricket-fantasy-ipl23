## Statistic(s) 

#### Batting level df 

```python

print(alldf.columns.values.tolist())
```

```python

groupbycols = ['date', 'city', 'inn_bat']

#bat_grouped_df = master_df.groupby(['bat_tm', 'batnum', 'bat', 'm.num'])
alldf_bats = alldf[ (alldf['b.num'] != 0.0) & ~(alldf['inning'].isnull()) ]

grnd_bat_df = alldf_bats.rename(columns = {'inning':'inn_bat','tm_bow':'vs_team'}).groupby(groupbycols) # batnum unique as the m.num per player


def custom_agg(group):
    data = {
        # Batting-Stats
        'tbinn': group['date'].nunique(),
        
        'tr': group['tr'].sum(),
        'tr_pp': (group['fpp'] * group['tr']).sum(),
        'tr_mid': (group['fmid'] * group['tr']).sum(),
        'tr_dth': (group['fdth'] * group['tr']).sum(),
        
        'tbov': group['m.tbov'].nunique(),
        'tbov.spin': group[group['bowlcat'] == 'Spin']['m.tbov'].nunique(),
        'tbov.pace': group[group['bowlcat'] == 'Pace']['m.tbov'].nunique(),
        
        'tout': (group['wkt']).sum(),
        'tout.spin': (group[group['bowlcat'] == 'Spin']['wkt']).sum(),
        'tout.pace': (group[group['bowlcat'] == 'Pace']['wkt']).sum()

        #'tbf': group['b.num'].count(),
        #'tbf_pp': group['fpp'].sum(),
        #'tbf_mid': group['fmid'].sum(),
        #'tbf_dth': group['fdth'].sum(),        

    }
    return pd.Series(data)

grnd_bat_level_df = grnd_bat_df.apply(custom_agg)

# Reset the index so that the primary key columns become regular columns
grnd_bat_level_df.reset_index(inplace=True)
```

```python

print(grnd_bat_level_df.shape)
print()
print(grnd_bat_level_df.columns.values.tolist())
```

```python

display(grnd_bat_level_df.head(5))
```

```python

print(grnd_bat_level_df.shape)

grouped_df_citydate = grnd_bat_level_df.groupby(['city','date']).sum().reset_index()
grouped_df_citydate['inn_bat'] = 'combined'

grouped_df_city = grnd_bat_level_df.groupby(['city']).sum().reset_index()
grouped_df_city['date'] = '9999-09-09'
grouped_df_city['inn_bat'] = 'combined'

grnd_bat_level_df_master = pd.concat([grnd_bat_level_df, grouped_df_citydate, grouped_df_city])

print(grnd_bat_level_df_master.shape)
print(grnd_bat_level_df_master.columns.values.tolist())

display(grnd_bat_level_df_master.tail(25))
```

#### Bowling level df 

```python

print(alldf.columns.values.tolist())
```

```python

bowgroupbycols = ['date', 'city', 'inn_bow']

#bat_grouped_df = master_df.groupby(['bat_tm', 'batnum', 'bat', 'm.num'])
alldf_bow = alldf[ (alldf['b.num'] != 0.0) & ~(alldf['inning'].isnull()) ]

grnd_bow_df = alldf_bow.rename(columns = {'inning':'inn_bow'}).groupby(bowgroupbycols) # batnum unique as the m.num per player

def custom_agg(group):
    data = {
        # Bowling-Stats        
        'trc': group['tr'].sum(),
        #'trc_pp': (group['fpp'] * group['tr']).sum(),
        #'trc_mid': (group['fmid'] * group['tr']).sum(),
        #'trc_dth': (group['fdth'] * group['tr']).sum(),
        
        #'tbwkt': (group['wkt']).sum(),
        #'tbwkt_pp': (group['wkt'] * group['fpp']).sum(),
        #'tbwkt_mid': (group['wkt'] * group['fmid']).sum(),
        #'tbwkt_dth': (group['wkt'] * group['fdth']).sum(),
        
        #'tbov': group['over'].nunique(),
        #'tbov_pp': (group[group['fpp'] == 1]['over']).nunique(),
        #'tbov_mid': (group[group['fmid'] == 1]['over']).nunique(),
        #'tbov_dth': (group[group['fdth'] == 1]['over']).nunique()
    }
    return pd.Series(data)

grnd_bow_level_df = grnd_bow_df.apply(custom_agg)

# Reset the index so that the primary key columns become regular columns
grnd_bow_level_df.reset_index(inplace=True)
```

```python

print(grnd_bow_level_df.columns.values.tolist())
```

```python

display(grnd_bow_level_df.head(5))
```

```python

print(grnd_bow_level_df.shape)

grouped_bow_citydate = grnd_bow_level_df.groupby(['city','date']).sum().reset_index()
grouped_bow_citydate['inn_bow'] = 'combined'

grouped_bow_city = grnd_bow_level_df.groupby(['city']).sum().reset_index()
grouped_bow_city['date'] = '9999-09-09'
grouped_bow_city['inn_bow'] = 'combined'

grnd_bow_level_df_master = pd.concat([grnd_bow_level_df, grouped_bow_citydate, grouped_bow_city])

print(grnd_bow_level_df_master.shape)
print(grnd_bow_level_df_master.columns.values.tolist())

display(grnd_bow_level_df_master.head(25))
```
