### 1. Get the 36 teams with all the CVC combinations of the Form Team 

```python

# Convert 'Form Team' column to string data type
df['Form Team'] = df['Form Team'].fillna('')
#df['Form Team'] = df['Form Team'].astype(str).replace('nan','')

# Sort the dataframe by 'Avg' column in descending order and select the top 6 players
top_six = df[ df['Form Team'] != ''].sort_values(by='Avg', ascending=False).head(6)

team_number = 1


# Iterate through all possible captain and vice-captain combinations
for i in range(len(top_six)):
    captain = top_six.iloc[i]
    
    for j in range(len(top_six)):
        if i == j:
            continue

        vice_captain = top_six.iloc[j]
        team_number += 1

        # Create new column name for the combination
        col_name = f"FT{team_number}"

        # Create a new column for the combination, copying the initial 'FT1' column
        df[col_name] = df['Form Team']

        # Update the new column for captain and vice-captain
        df.loc[captain.name, col_name] += " + c"
        df.loc[vice_captain.name, col_name] += " + vc"
        
print("That's a wrap")
```

```python

# Filter the top_six dataframe to include only the players in the 'Form Team' column
form_team = df[df['Form Team'] != '']
team_number = 1

# Iterate through all players in the 'Form Team' column
for i in range(len(form_team)):
    captain = form_team.iloc[i]

    # Remove captain from the list of possible vice-captains
    possible_vice_captains = form_team[form_team['Name'] != captain['Name']]

    # Randomly select a vice-captain from the remaining players
    vice_captain = possible_vice_captains.sample()

    # Create new column name for the combination
    col_name = f"FT{team_number}"

    # Create a new column for the combination, copying the initial 'FT1' column
    df[col_name] = df['Form Team']

    # Update the new column for captain and vice-captain
    df.loc[captain.name, col_name] += " + c"
    df.loc[vice_captain.index[0], col_name] += " + vc"
    
    team_number += 1
```
