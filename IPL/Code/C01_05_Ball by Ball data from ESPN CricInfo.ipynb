```python

# Get the HTML content of the website
response = requests.get('https://www.espncricinfo.com/series/indian-premier-league-2023-1345038/chennai-super-kings-vs-kolkata-knight-riders-61st-match-1359535/ball-by-ball-commentary')

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, 'html.parser')
soup_innings = BeautifulSoup(response.content, 'html5lib')

# pprint soup to see what is going on here
pprint(soup_innings)
```

```python

# Find all the ball-by-ball commentary elements
commentary_elements = soup.find_all('div', class_='commentary')

pprint(commentary_elements)
```

```python
# Extract the data from the commentary elements
data = []
for commentary_element in commentary_elements:
    # Extract the over number
    over_number = commentary_element.find('span', class_='over').text

    # Extract the ball number
    ball_number = commentary_element.find('span', class_='ball').text

    # Extract the batsman
    batsman = commentary_element.find('span', class_='batsman').text

    # Extract the bowler
    bowler = commentary_element.find('span', class_='bowler').text

    # Extract the delivery type
    delivery_type = commentary_element.find('span', class_='delivery-type').text

    # Extract the runs scored
    runs_scored = commentary_element.find('span', class_='runs').text

    # Extract the wickets taken
    wickets_taken = commentary_element.find('span', class_='wickets').text

    # Extract the extras
    extras = commentary_element.find('span', class_='extras').text

    # Extract the total runs
    total_runs = commentary_element.find('span', class_='total-runs').text

    # Append the data to the list
    data.append({
        'over_number': over_number,
        'ball_number': ball_number,
        'batsman': batsman,
        'bowler': bowler,
        'delivery_type': delivery_type,
        'runs_scored': runs_scored,
        'wickets_taken': wickets_taken,
        'extras': extras,
        'total_runs': total_runs
    })

```

```python


# Print the data
print(data)

```
