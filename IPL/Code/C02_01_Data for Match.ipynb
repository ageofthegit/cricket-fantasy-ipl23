```python


df_match_curr = df_masterfile.loc[ df_masterfile['team_nm'].isin( [ team_name_a, team_name_b] ) ]

print(df_match_curr.shape)
print(df_match_curr.columns.values.tolist())

```

### 1 Match Summary

```python

display(df_match_curr)
```

```python

df_player_rup = df_match_curr.groupby(['team_nm','name'])\
    .agg({'mnum':'nunique', 'bat_pts':'sum', 'rs':'sum', 'four':'sum', 'six':'sum', 'bowf':'count', 'bow_pts':'sum','wkt':'sum','inn_bow':'count'})\
        .rename(columns = {'mnum':'named_in_squad','bowf':'bat_inn','inn_bow':'bow_inn'})\
            .reset_index()

print(df_player_rup.shape)
print()
print(df_player_rup.columns.values.tolist())
print()
```

```python

df_player_rup.groupby('team_nm').agg({'name':'count'})
```

```python

display(df_player_rup[df_player_rup.team_nm.isin( [team_name_a]) ].head(25))
```

### 1.2 Match Summary playing first

```python

df_player_inn1_rup = df_match_curr[df_match_curr.inn_bat == 1].groupby(['team_nm','name'])\
    .agg({'mnum':'nunique', 'bat_pts':'sum', 'rs':'sum', 'four':'sum', 'six':'sum', 'bowf':'count', 'bow_pts':'sum','wkt':'sum','inn_bow':'count'})\
        .rename(columns = {'mnum':'i1_named_in_squad','bat_pts':'i1_bat_pts'\
                           , 'rs':'i1_rs', 'four':'i1_four', 'six':'i1_six' , 'bowf':'i1_bowf', 'bow_pts':'i2_bow_pts','wkt':'i2_wkt', 'inn_bow':'i2_inn'})\
            .reset_index()

print(df_player_inn1_rup.shape)
print()
print(df_player_inn1_rup.columns.values.tolist())
print()
```

```python

display(df_player_inn1_rup.head(25))
```

### 1.3 Innings 2 Roll-Up

```python

df_player_inn2_rup = df_match_curr[df_match_curr.inn_bat == 2].groupby(['team_nm','name'])\
    .agg({'mnum':'nunique', 'bat_pts':'sum', 'rs':'sum', 'four':'sum', 'six':'sum', 'bowf':'count', 'bow_pts':'sum','wkt':'sum','inn_bow':'count'})\
        .rename(columns = {'mnum':'i2_named_in_squad','bat_pts':'i2_bat_pts'\
                           , 'rs':'i2_rs', 'four':'i2_four', 'six':'i2_six' , 'bowf':'i2_bowf', 'bow_pts':'i1_bow_pts','wkt':'i1_wkt', 'inn_bow':'i1_inn'})\
            .reset_index()

print(df_player_inn2_rup.shape)
print()
print(df_player_inn2_rup.columns.values.tolist())
print()
```

```python

display(df_player_inn1_rup.head(25))
```


### 2. Maximum Pts. 


```python

df_player_max = df_match_curr.groupby(['team_nm','name'])\
    .agg({'bat_pts':'max', 'rs':'max', 'bow_pts':'max','wkt':'max'})\
        .rename(columns = {'bat_pts':'max_bat_pts','rs':'max_rs','bow_pts':'max_bow_pts','wkt':'max_wkt'})\
            .reset_index()

print(df_player_max.shape)
print(df_player_max.columns.values.tolist())
```

```python

display(df_player_max[df_player_max.team_nm.isin( ['Rajasthan Royals']) ].head(25))
```

### 3. Merging the datasets 

```python

"""
df_mg = pd.merge(df_player_rup, df_player_max, df_player_inn1_rup, , on = ['team_nm','name'], how = 'outer')

print(df_mg.shape)
print()
print(df_mg.columns.values.tolist())
print()
"""
```

```python

df_list = [ df_player_max, df_player_rup, df_player_inn1_rup, df_player_inn2_rup ]

def merge_dataframes(left, right):
    return pd.merge(left, right, on = ['team_nm','name'], how='outer')

df_mg = reduce(merge_dataframes, df_list)
```

```python

print(df_mg.shape)
print()
print(df_mg.columns.values.tolist())
print()
```

```python

display(df_mg.head())
```

```python

### EXPORT 

df_exp = df_mg[ df_mg.team_nm.isin( [ team_name_a, team_name_b ] ) ]

print(df_exp.shape)
print(df_exp.columns.values.tolist())
```
